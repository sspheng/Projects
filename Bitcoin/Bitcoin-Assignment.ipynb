```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/Users/sophengsok/Documents/DataScience/!Dataset/bitcoindata.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.Timestamp=pd.to_datetime(df.Timestamp, unit='s')
```

```python
df
```

```python
df.index=df.Timestamp
```

```python
df
```

```python
df=df.resample('D').mean()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Forwardfill']=df['Weighted_Price'].ffill()
```

```python
df
```

```python
df['Movingaverage']=df['Forwardfill'].rolling(window=4).mean()
```

```python
df
```

```python
df['Price']=np.where(df['Weighted_Price'].isna(),df['Movingaverage'],df['Weighted_Price'])
```

```python
df
```

```python
data=df['Price']
```

```python
data
```

```python
import statsmodels.api as sm
plt.figure(figsize=[20,10])
decomp=sm.tsa.seasonal_decompose(data.head(100))
fig=decomp.plot()
fig.set_size_inches(20,10)
fig.tight_layout()
plt.show()
```
